```python
import torch.nn as nn
import torch.nn.functional as F

conv1 = nn.Conv2d(1, 64, 3,padding=1)
conv2 = nn.Conv2d(64,128,3,padding=1)
conv3 = nn.Conv2d(128,512,3,padding=1)
pool = nn.MaxPool2d(2,2)
fc1 = nn.Linear(512*26*40, 500)
fc2 = nn.Linear(500, 50)
dropout = nn.Dropout(0.25)
```

```python
gray_img.shape
```

```python
x = torch.tensor(gray_img)
x = x.unsqueeze(0).unsqueeze(0)
x.shape
```

```python
# helper function for visualizing the output of a given layer
# default number of filters is 4
def viz_layer(layer, n_filters= 4):
    fig = plt.figure(figsize=(20, 20))
    
    for i in range(n_filters):
        ax = fig.add_subplot(1, n_filters, i+1, xticks=[], yticks=[])
        # grab layer outputs
        ax.imshow(np.squeeze(layer[0,i].data.numpy()), cmap='gray')
        ax.set_title('Output %s' % str(i+1))
```

```python
x = pool(F.relu(conv1(x)))
viz_layer(x)
```

```python
viz_layer(x)
```

```python
x = pool(F.relu(conv2(x)))
viz_layer(x)
```

```python
x = pool(F.relu(conv3(x)))
viz_layer(x)
```
